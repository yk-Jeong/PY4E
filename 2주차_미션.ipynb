####Q1. 가위바위보 게임 
- 조건 1. 함수의 인자로는 **나의 가위바위보 선택**이 들어감
- 조건 2. 누가 무엇을 냈고, 누가 승리했는지 출력되어야 함

```python
import random 


def rcp(my):
  my = input("가위 바위 보 중 하나를 골라주세요. ")
  computer = random.randint(0, 2)
  winner = "무승부."
# 컴퓨터의 입력을 문자열로 변환
  if computer == 0: 
    computer = '가위'
  elif computer == 1:
    computer = '바위'
  else:
    computer = '보'

# 승패 구분
  if my == computer:
    pass
  elif (my == '가위' and computer == '보') or (my == '바위' and computer == '가위') or (my == '보' and computer == '바위'):
    winner = "유저 승!"
  else: 
    winner = "컴퓨터 승!"

  return print(f'\n나: {my} \n컴퓨터: {computer}\n{winner}')

rcp(my)
```

####Q2. 연봉 계산기
다음의 세율표를 토대로 세전, 세후 연봉을 함께 출력해 보자.

image.png

```python
def payment(month):
  month = int(input("월급을 입력해주세요. (단위: 만원, e.g. 300만원 -> 300 입력) "))
  before = month * 12
  if before <= 1200: 
    after = before * 0.94
  elif before > 1200 and before <= 4600:
    after = before * 0.85
  elif before > 4600 and before <= 8800:
    after = before * 0.76
  elif before > 8800 and before <= 15000:
    after = before * 0.65
  elif before > 15000 and before <= 30000:
    after = before * 0.62
  elif before > 30000 and before <= 50000:
    after = before * 0.60
  else:
    after = before * 0.58
  return print(f'\n세전 연봉: {before}만원 \n세후 연봉: {after}만원')

payment(month)
```

####Q3. 학점 변환기

다음의 학점 환산표를 토대로 학생의 이름과 원점수, 학점을 출력해 보자.
image.png

```python
def grader(name, point):
  name = input("이름을 입력하세요. ")
  point = int(input("원점수를 입력하세요. "))
  grade = 'A+'
  if point >= 95 and point <= 100:
    pass
  elif point >= 90 and point <= 94:
    grade = 'A'
  elif point >= 85 and point <= 89:
    grade = 'B+'
  elif point >= 80 and point <= 84:
    grade = 'B'
  elif point >= 75 and point <= 79:
    grade = 'C+'
  elif point >= 70 and point <= 74:
    grade = 'C'
  elif point >= 65 and point <= 69:
    grade = 'D+'
  elif point >= 60 and point <= 64:
    grade = 'D'
  else:
    grade = 'F'
  return print(f'\n학생이름: {name}\n원점수: {point}점 \n학점: {grade}')

grader(name, point)
```
